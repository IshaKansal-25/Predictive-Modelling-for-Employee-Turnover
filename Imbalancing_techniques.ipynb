### Plotting Histograms

```python
# defining a function to view distribution of numeric features
def plot_hist(col,data=data):
    plt.figure(figsize=(10,15))
    plt.subplot(2,1,1)
    sns.histplot(data=data[data['Attrition'] == 0],x=col, bins=10)
    plt.title(f'Distribution of {col} Stayed')
    plt.subplot(2,1,2)
    sns.histplot(data=data[data['Attrition'] == 1],x=col, bins=10)
    plt.title(f'Distribution of {col} Attrited')
    plt.plot()
```

### Plotting categorical values using countplot

```python
def plot_count(col,data=data):
    sns.countplot(x=col, data=data,hue='Attrition')
    plt.xticks(rotation=60)
    plt.title(f'Distribution of {col}')
    plt.show()
```

Converting categorical data to numerical data using **get_dummies()**

```python
dummy_col = [column for column in data.drop('Attrition', axis=1).columns if data[column].nunique() < 20]
converted_data = pd.get_dummies(data, columns=dummy_col, drop_first=True, dtype='uint8')
converted_data.info()
```

```python
X = converted_data.drop('Attrition', axis=1)
y = converted_data['Attrition']
y.value_counts()
```

The data is imbalanced

```python
converted_data.columns
```

# SMOTE


```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)
y_resampled.value_counts()
```

## Splitting the datset into training and testing sets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

## Standard scaling numeric values

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = pd.DataFrame(scaler.fit_transform(X_train), columns = X_train.columns)
X_test= pd.DataFrame(scaler.transform(X_test), columns = X_test.columns)

X_train.head()
```

## Applying Machine Learning Models

importing libraries for evaluating models

```python
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, cohen_kappa_score, precision_score, recall_score, f1_score, roc_curve, RocCurveDisplay

import math
def gmean(confusion):
  TP = confusion[1,1] # true positive
  TN = confusion[0,0] # true negatives
  FP = confusion[0,1] # false positives
  FN = confusion[1,0] # false negatives

  senstivity= TP / float(TP+FN)
  specificity= TN / float(TN+FP)
  gmean=math.sqrt( senstivity * specificity)
  print('Geometric Mean : ', gmean)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logistic= LogisticRegression(solver='liblinear', penalty='l1')
logistic.fit(X_train, y_train)
```

Evaluation of model :

```python
logistic_predict=logistic.predict(X_test)
```

```python
print('Cohen Kappa Score: ', cohen_kappa_score(y_test, logistic_predict))
```

**Since Cohen Kappa Score is lying in range 0.81-1.00                           -> Therefore there is *Perfect Agreement***

```python
print('Accuracy score:',accuracy_score(y_test, logistic_predict))
```

```python
print('Precision score:',precision_score(y_test, logistic_predict))
```

```python
print('Recall score:',recall_score(y_test, logistic_predict))
```

```python
print('F1-score:',f1_score(y_test, logistic_predict))
```

```python
roc_auc=roc_auc_score(y_test,logistic_predict)
print('ROC AUC Score: ', roc_auc)
```

```python
fpr, tpr, thresholds = roc_curve(y_test,logistic_predict)

display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc,estimator_name='Logistic Regression')
display.plot()
plt.show()
```

```python
confusion = confusion_matrix(y_test,logistic_predict)
sns.heatmap(confusion, annot=True)
plt.show()
```

```python
print(confusion)
```

```python
gmean(confusion)
```

## Applying Machine Learning Models

importing libraries for evaluating models

```python
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, cohen_kappa_score, precision_score, recall_score, f1_score, roc_curve, RocCurveDisplay

import math
def gmean(confusion):
  TP = confusion[1,1] # true positive
  TN = confusion[0,0] # true negatives
  FP = confusion[0,1] # false positives
  FN = confusion[1,0] # false negatives

  senstivity= TP / float(TP+FN)
  specificity= TN / float(TN+FP)
  gmean=math.sqrt( senstivity * specificity)
  print('Geometric Mean : ', gmean)
```
